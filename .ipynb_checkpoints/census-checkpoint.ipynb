```python
# 3. For each state, what percentage of total workers belong to each ethnicity?
eth_group_count = df_business_summary.groupby('NAME')['ETH_GROUP_LABEL'].value_counts()
eth_group_count.unstack().fillna(0)
```

```python
# store the total row in a variable
states_total = eth_group_count.unstack()["Total"]
# calculate the percentage
eth_group_count.unstack().fillna(0).div(states_total, axis=0).mul(100).round(2)
```

```python
# plot the data as bar chart
remove_total = eth_group_count.unstack().fillna(0).div(states_total, axis=0).mul(100).round(2).drop('Total', axis=1)
remove_total.plot(kind='bar', stacked=True, figsize=(20,10))
plt.show()
```

```python
# 4. For each state, what is the total number of business and how much average revenue have they generated for 2018?
# total number of business
df_business_summary['FIRMPDEMP'] = df_business_summary['FIRMPDEMP'].astype(float)
df_business_summary.groupby('NAME')['FIRMPDEMP'].sum()
```

```python
# plot the data
df_business_summary.groupby('NAME')['FIRMPDEMP'].sum().plot(kind='bar', figsize=(20,10))
plt.show()
```

```python
# 5. The percentage of male and female employees for each state.
gender_group_count = df_business_summary["SEX_LABEL"].value_counts()
gender_group_count
```

```python
# group the number males and females per state
gender_group_count = df_business_summary.groupby('SEX_LABEL')['NAME'].value_counts()
gender_group_count = gender_group_count.unstack().fillna(0).T
male_count = gender_group_count['Male'] + gender_group_count['Equally male/female']
female_count = gender_group_count['Female'] + gender_group_count['Equally male/female']
```

```python
percent_males = male_count/gender_group_count['Total']
percent_females = female_count/gender_group_count['Total']
print(f"The percentage of males is: {round(percent_males.mean()*100,2)}%")
print(f"The percentage of females is: {round(percent_females.mean()*100,2)}%")
```

```python
# plot the data as pie chart
labels = 'Male','Female'
sizes = [percent_males.mean(), percent_females.mean()]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# 6. For each state, the average number of years business have been in business.
df_business_summary['YIBSZFI'].value_counts()
```

```python
df_business_summary['YIBSZFI_LABEL'].value_counts()
```

```python
# replace the codes with the actual values
# 311 Firms with less than 2 years in business
# 318 Firms with 2 to 3 years in business
# 319 Firms with 4 to 5 years in business
# 321 Firms with 6 to 10 years in business
# 322 Firms with 11 to 15 years in business
# 323 Firms with 16 or more years in business
codes = {
    '311': 1,
    '318': 2.5,
    '319': 4.5,
    '321': 8,
    '322': 13,
    '323': 16,
    '001': 'Total'
}
df_years_in_business = df_business_summary.replace({'YIBSZFI': codes})
# remove the total row
df_years_in_business = df_years_in_business[df_years_in_business['YIBSZFI'] != 'Total']
df_years_in_business.groupby('NAME')['YIBSZFI'].mean()
```

```python
# plot the data as bar chart showing the average number of years in business
df_years_in_business.groupby('NAME')['YIBSZFI'].mean().plot(kind='bar', figsize=(20,5))
plt.show()
```

```python
# generate heatmap showing correlation between the data to see how company size, company revenue, employee type (gender, veterans, etc.), company location and company years active are correlated
corr = df_business_summary.corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(data = corr,
annot = True,
mask = mask,
cmap = 'PuBuGn_r',
vmin = -1,
vmax = 1)
```
